###Uncertainty:

```python
results = []
for i in range(1000):
    results.append(model.run(params={'characteristic_time':np.random.normal(15,3)})['teacup_temperature'])

resultsdf = pd.DataFrame(results)
resultsdf.head()
```

```python
for i, row in resultsdf.iterrows():
    plt.plot(row.index, row.values, 'b', alpha=.02)
```
